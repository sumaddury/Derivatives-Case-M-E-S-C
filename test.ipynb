```python
exp_day = "2022-05-20"
day_before = "2022-05-19"

# Get difference between exp day high vs prev day low and vice versa
exp_day_high = df[df["Date"] == exp_day]["High"].values[0]
exp_day_low = df[df["Date"] == exp_day]["Low"].values[0]
prev_day_high = df[df["Date"] == day_before]["High"].values[0]
prev_day_low = df[df["Date"] == day_before]["Low"].values[0]
print(exp_day_high - prev_day_low)
print(prev_day_high - exp_day_low)
```
